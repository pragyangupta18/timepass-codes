```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Your data
density_particle = np.array([1500, 1600, 1700, 1800, 1900])
density_fluid = np.array([1000, 1000, 1000, 1000, 1000])
viscosity = np.array([0.001, 0.001, 0.001, 0.001, 0.001])
draft = np.array([0.05, 0.06, 0.07, 0.08, 0.091])

# Stack your data as you did before
X = np.column_stack((density_particle, density_fluid, viscosity))
y = draft

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Input values for prediction
new_density_particle = float(input("Enter density of particle (kg/m^3): "))
new_density_fluid = float(input("Enter density of fluid (kg/m^3): "))
new_viscosity = float(input("Enter viscosity (kg/(m*s)): "))
new_input = np.array([new_density_particle, new_density_fluid, new_viscosity]).reshape(1, -1)

# Predict the draft
predicted_draft = model.predict(new_input)

# Create a graph to visualize all data points and the predicted value
plt.scatter(density_particle, draft, label='Data', color='blue')
plt.scatter(new_density_particle, predicted_draft, label='Predicted', color='red', marker='x', s=100)
plt.xlabel('Density Particle (kg/m^3)')
plt.ylabel('Draft')
plt.legend()

# Display the predicted value as text on the plot
plt.text(new_density_particle, predicted_draft, f'Predicted: {predicted_draft[0]:.2f}', ha='right', va='bottom')

plt.title('Draft vs. Density Particle')
plt.grid(True)
plt.show()
```
