House Price Prediction expert system

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train is the size of house in 1000 sqft
#y_train is the price of house in 1000 dollars
x_train = np.array([2.1, 1.4, 1.5, 0.8, 3.2])
y_train = np.array([400, 232, 315, 178, 870])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

Numpy parameter .shape is return a python tuple with an entry for each dimention (size of array in tuple form)
x_train.shape[0] gives the length of array

```python
# m is number of training example
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples  : {m}")
```

```python
i=0 #change it as 1,2...to see more
x_i = x_train[i]
y_i = y_train[i]
print(f"x^({i}), y^({i}) = ({x_i}, {y_i})")
```

```python
#Plot the data points
plt.scatter(x_train, y_train, marker='X', c='r', label='Actual Values')
plt.title("Housing Prices")
plt.xlabel('Size(in 1000sqft)')
plt.ylabel('Price(in $1000)')
plt.show()
```

```python
w=100
b=100
print(f"w : {w}")
print(f"b : {b}")
```

```python
def compute_model_output(x, w, b):
  """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples
      w,b (scalar)    : model parameters
    Returns
      f_wb (ndarray (m,)): model prediction
    """
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range (m):
    f_wb[i] = w * x[i] + b
  return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
plt.plot(x_train, tmp_f_wb, c='b', label='Predicted Price')
plt.scatter(x_train, y_train, marker='X', c='r', label='Actual values')
plt.title("House Price Prediction System")
plt.xlabel('Size(in 1000sqft)')
plt.ylabel('Price(in $1000)')
plt.legend()
plt.show()
```
